### Canvas Plot - Mupltiple Graph - in one single frame
1. If want to view and compare multiple parameters at single place we can use plt.subplot()
2. where we can specify the plot size and figure size(row, col) and at runtime.

```python
# first create one script which will create seperate subplot with row*col matrix
# Lets compare all Col data which ha SF data
sf_col=[]
for column in housing_data.columns:
    if 'SF' in column:
        sf_col.append(column)
sf_col
```

```python
## here total 9 columns with SF so we need to create script which will run 
fig, axs= plt.subplots(nrows=3, ncols=3, figsize=(20,20))
# fig - figure size - 20*20
# axs - [row, col]

for i in range(0, len(sf_col)):
    row= i // 3 # it will return only quotients
    col= i % 3 # it will return only reminder
    ax=axs[row, col]
    plot=sns.scatterplot(x=sf_col[i], y='SalePrice', data=housing_data, ax=ax)
```

### Categorising data for better visualization
1. Adding new columns in dataset based on existing data
2. Apply condition to create new data.

```python
housing_data['YearBuilt'].value_counts() 
```

```python
# here we are using np.where() where we are 
housing_data['House_type']=np.where(housing_data['YearBuilt'] > 2000, 'New', 'Old') 
housing_data['House_type'].value_counts() 
```

```python
# plot Box graph of House_type
sns.boxplot(x='House_type', data=housing_data)
```

```python
sns.countplot(x='House_type', data=housing_data, palette='bright')
```

```python
# above plot showing which are my new annd old houses
# but if i want  to know which Exterior1st material used in this new and old houses,
# here we can use 'hue'='Exterior1st' where it will show exactly which material use in new and old
sns.countplot(x='House_type',hue='Exterior1st',data=housing_data,palette='bright')
```

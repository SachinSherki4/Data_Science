```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

### Pie chart representatio of Data

```python
movie_data=pd.read_csv('../../Data_Files/movie_metadata.csv')
movie_data.shape
```

### represent data in pie - percentage form

```python
movie_data.columns
```

```python
movie_score=movie_data['imdb_score'].value_counts()
```

```python
plt.pie(movie_score, labels=movie_score.index)
```

```python
# dividing scores into different types
score_bins=[0,4,6,8,10]
score_labels=['Love It ', 'Liked It ','It Was Okay ',"Didnt Like It "]
movie_data['Movie Ratings']=pd.cut(movie_data['imdb_score'], bins=score_bins, labels=score_labels)
```

```python
ratings=movie_data['Movie Ratings'].value_counts()
plt.pie(ratings, labels=ratings.index, autopct="%1.1f%%")
```
